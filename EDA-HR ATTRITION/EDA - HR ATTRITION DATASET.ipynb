# DATA SCIENCE PROJECT
### EXPLORATORY DATA ANALYSIS OF EMPLOYEE ATTRITION; THE HR ATTRITION DATASET


An organization’s ability to survive in our globally competitive business climate is largely dependent on several factors. A committed workforce is one of those factors. That is why there must be cause for worry when there is a very high attrition rate among a company’s labor force.

On the face of it, employee attrition can happen for several reasons such as unhappiness about employee benefits or the pay structure, a lack of employee development opportunities, and even poor conditions in the workplace. The question however is; what are the key factors that influence employee attrition.

My goal is to examine the data set and explore the dataset and ascertain the various relationships that exist between all the various factors, thereby helping the appropriate executives take the necessary steps to retain the employee.


The Dataset
In this project, I perform an analysis on a publicly available dataset (https://www.kaggle.com/datasets/itssuru/hr-employee-attrition?select=HR-Employee-Attrition.csv) 

The key steps covered in this project are as follows;
1.	Importing the relevant libraries
2.	Cleaning the data
3.  Exploratory Data Analysis
6.	Concluding remarks.
